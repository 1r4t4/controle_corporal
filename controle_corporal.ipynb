```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
controle_corporal = pd.read_csv('controle_corporal.csv')
controle_corporal
```

```python
# Extraindo dados para variáveis
peso = controle_corporal.iloc[0, :].values[1:]
gordura_corporal = controle_corporal.iloc[2, :].values[1:]
porcentagem_musculos = controle_corporal.loc[3].values[1:]
metabolismo = controle_corporal.loc[4].values[1:]
idade_metabolica = controle_corporal.loc[5].values[1:]
gordura_visceral = controle_corporal.loc[6].values[1:]
massa_magra = controle_corporal.loc[7].values[1:]


# Obter as datas a partir da primeira linha
datas = controle_corporal.columns[1:]
```

```python
# Plotar os gráficos separados
plt.plot(datas, peso, marker='o')
plt.ylabel('Peso')

# Exibir os gráficos
plt.xlabel('Data')
plt.suptitle('Variação do peso com o tempo', fontsize=16)
plt.show()

peso = peso.astype(np.float)  # Converter para tipo numérico
```

```python

# Extrair as colunas de interesse, exceto a primeira linha
gordura_corporal = controle_corporal.iloc[2, :].values[1:]
gordura_corporal = gordura_corporal.astype(np.float)  # Converter para tipo numérico

# Obter as datas a partir da primeira linha
datas = controle_corporal.columns[1:]

# Plotar o gráfico para a gordura corporal
plt.plot(datas, gordura_corporal, marker='o')
plt.ylabel('Gordura Corporal')

# Ajustar os limites dos eixos y
plt.ylim(np.min(gordura_corporal) - 1, np.max(gordura_corporal) + 1)

# Exibir o gráfico
plt.xlabel('Data')
plt.title('Análise da Gordura Corporal')
plt.show()
```

```python
# Extrair as colunas de interesse, exceto a primeira linha
porcentagem = controle_corporal.iloc[3, :].values[1:]
porcentagem = np.vectorize(lambda x: x.replace(',', '.'))(porcentagem)  # Substituir vírgulas por pontos
porcentagem = porcentagem.astype(np.float)  # Converter para tipo numérico

# Obter as datas a partir da primeira linha
datas = controle_corporal.columns[1:]

# Plotar o gráfico para a gordura corporal
plt.plot(datas, porcentagem, marker='o')
plt.ylabel('Porcentagem')

# Ajustar os limites dos eixos y
plt.ylim(np.min(porcentagem) - 1, np.max(porcentagem) + 1)

# Exibir o gráfico
plt.xlabel('Data')
plt.title('Análise da Porcentagem de Músculos')
plt.show()
```

```python

# Extrair as colunas de interesse, exceto a primeira linha
imc = controle_corporal.iloc[1, :].values[1:]
imc = imc.astype(np.float)  # Converter para tipo numérico

# Obter as datas a partir da primeira linha
datas = controle_corporal.columns[1:]

# Plotar o gráfico para a gordura corporal
plt.plot(datas, imc, marker='o')
plt.ylabel('IMC')

# Ajustar os limites dos eixos y
plt.ylim(np.min(imc) - 1, np.max(imc) + 1)

# Exibir o gráfico
plt.xlabel('Data')
plt.title('Análise do IMC')
plt.show()
```

```python
#obtendo os dados no formato object
massa_magra = controle_corporal.iloc[7, :].values[1:]

massa_magra = [float(valor.replace(',', '.')) for valor in massa_magra]

# Plotar os gráficos separados
plt.plot(datas, massa_magra, marker='o')
plt.ylabel('Massa Magra')

# Exibir os gráficos
plt.xlabel('Data')
plt.suptitle('Variação da massa magra com o tempo', fontsize=16)
plt.show()

#for i in massa_magra:
#  float(massa_magra[i].replace(',', '.'))
# massa_magra = massa_magra.astype(np.float)  # Converter para tipo numérico
```
